```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop('Attrition_Flag',axis=1)
y = df['Attrition_Flag']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train.head()
```

```python
X_train.columns
```

```python
X_train_scaled = scaler.fit_transform(X_train[['Customer_Age','Credit_Limit', 'Total_Revolving_Bal','Total_Trans_Amt']])
```

```python
X_train_scaled = pd.DataFrame(X_train_scaled,columns=['Customer_Age','Credit_Limit', 'Total_Revolving_Bal','Total_Trans_Amt'])
X_train_scaled.head()
```

```python
X_train_rem = X_train.drop(['Customer_Age','Credit_Limit', 'Total_Revolving_Bal','Total_Trans_Amt'],axis=1)
X_train_rem.head()
```

```python
X_train_rem.index = np.arange(0,7595)
```

```python
X_train= pd.concat([X_train_scaled,X_train_rem],axis=1)
X_train.head()
```

```python
X_test_scaled = scaler.transform(X_test[['Customer_Age','Credit_Limit','Total_Revolving_Bal','Total_Trans_Amt']])
X_test_scaled = pd.DataFrame(X_test_scaled,columns=['Customer_Age','Credit_Limit', 'Total_Revolving_Bal','Total_Trans_Amt'])
X_test_scaled.head()
```

```python
X_test_rem = X_test.drop(['Customer_Age','Credit_Limit', 'Total_Revolving_Bal','Total_Trans_Amt'],axis=1)
X_test_rem.index = np.arange(0,2532)
X_test= pd.concat([X_test_scaled,X_test_rem],axis=1)
X_test.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression(random_state=42)
```

```python
log.fit(X_train,y_train)
```

```python
y_pred = log.predict(X_test)
```

```python
from sklearn.metrics import *
```

```python
print(confusion_matrix(y_pred=y_pred,y_true=y_test))
print("----------------------------------------------")
print(accuracy_score(y_pred=y_pred,y_true=y_test))
print("----------------------------------------------")
print(classification_report(y_pred=y_pred,y_true=y_test))
print("----------------------------------------------")
print(r2_score(y_pred=y_pred,y_true=y_test)**0.5)
```
